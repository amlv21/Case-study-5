# Check whether all the zones generate the same amount of sales.

Level of significance = 0.05

Null Hypothesis = All zones generate same amount of sales

Alternatw Hypothesis = Zones dont have same amount of sales

Taking one way test:
    
    Dependent variable:Amount of sales
    Independent variable:ZONE A,ZONE B,ZONE C,ZONE D


```python
from scipy.stats import f_oneway
fvalue, pvalue = stats.f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'], df['Zone - D'])
```

```python
print(fvalue,pvalue)
```

CHECKING P VALUE WITH LEVEL OF SIGNIFICANCE

```python
if pvalue<0.05:
   print('reject null hypothesis')
else:
    print('accept null hypothesis')
```

#### INSIGHTS:
          We reject null hypothesis since p-value less 0.05
          The amount of sales is different in different zones.
